```python
from copy import deepcopy
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')
```

```python

data = pd.read_csv('user_data.csv')
#data.head()
```

```python
from sklearn.manifold import TSNE
X_embedded = TSNE(n_components=2).fit_transform(data)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(4).fit(X_embedded)
```

```python

centers=kmeans.cluster_centers_
```

```python
color = {0: 'red', 1: 'blue', 2: 'green', 3: 'purple'}
colors = [color[i] for i in kmeans.labels_]
#kmeans.predict(data)
```

```python
plt.plot( color=colors)
import matplotlib.pyplot
matplotlib.pyplot.scatter(np.append(X_embedded[:,0], centers[:, 0]), 
                          np.append(X_embedded[:, 1], centers[:, 1]),
                          color=colors + ['black']*4)
```
